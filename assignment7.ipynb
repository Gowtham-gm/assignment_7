```python
print("The people who are married, working in private sector and having a masters degree are")

engine.execute('SELECT count(*) FROM sqladb WHERE education=" Masters" AND workclass=" Private" AND marital_status=" Married-civ-spouse";').fetchall()
```

```python
print(" The average, minimum and maximum age group for people working in different sectors  are")

engine.execute("SELECT occupation, Min(age), Max(age), Avg(age) FROM sqladb Group By occupation;").fetchall()
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data")
df
```

```python
df.columns = ["age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race",
             "sex","capital_gain","capital_loss","hours_per_week","native_country","income_class"]
df.head()
```

```python
from sqlalchemy import create_engine

engine = create_engine('sqlite://', echo=False)  # Creating engine

df.to_sql("sqladb",con=engine,chunksize=1, if_exists='replace')
```

```python
connection  = sqlite3.connect('sqladb')

cursor = connection.cursor()
```

```python
print("\n 1. Create an sqlalchemy engine using a sample from the data set\n",'-'*80, sep='')

engine.execute("SELECT * FROM sqladb Limit 5").fetchall()
```

```python
print("\n 2. First Update (Changing Private workclass to Updated)\n",'-'*80, sep='')

engine.execute('UPDATE sqladb SET marital_status = "Updated" WHERE marital_status = " Divorced"')
engine.execute("SELECT * FROM sqladb").fetchall()
```

```python
print("\n 2. Second Update (Changing native-country from United-States to America)\n",'-'*80, sep='')

engine.execute('UPDATE sqladb SET native_country = "America" WHERE native_country = " United-States"')
engine.execute("SELECT * FROM sqladb;").fetchall()
```

```python
print("\n 3. First delete query (Delete rows where education = Masters) \n",'-'*80, sep='')

engine.execute('DELETE FROM sqladb WHERE education = " Masters"')
engine.execute("SELECT * FROM sqladb").fetchall()
```

```python
print('\n 4. Write two filter queries \n','-'*35, sep='')

engine.execute("SELECT race, COUNT(*) AS Total FROM sqladb GROUP BY race;").fetchall()
```

```python
print("\n 5. First function query (Working hours as per Occupation) \n",'-'*65, sep='')

def workhrs():
    
    connection  = sqlite3.connect("sqladb")
    cursor = connection.cursor()
           
    return engine.execute('SELECT education, hours_per_week FROM sqladb GROUP BY education ').fetchall()

workhrs()
```

```python
 print("\n 5. Second function query (Average age for getting married) \n",'-'*65, sep='')

def maritalage():
    
    connection  = sqlite3.connect("sqladb")
    cursor = connection.cursor()
    
    return engine.execute('SELECT marital_status, avg(age) FROM sqladb GROUP BY marital_status ').fetchall()

maritalage()
```
